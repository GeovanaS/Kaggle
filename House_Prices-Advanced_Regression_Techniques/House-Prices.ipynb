### NaNs da coluna "LotFrontage":

É uma coluna numérica. Assim, podemos preencher esse valores de diversas formas, como utilizar o comando .fillna(); com média, moda, mediana dos valores que possuímos;  ou o KNN. Neste dataset, vamos usar o KNN.

#### Utilizando KNN:
O scikit-learn tem um imputador de valores de NaN que utiliza da média dos N vizinhos próximos, [encontrado neste link.](https://scikit-learn.org/stable/modules/generated/sklearn.impute.KNNImputer.html)

Ele tem alguns parametros básicos:

•*missing_value*:O tipo do valor que você quer  substituir.

•*n_neighborsint*: A quantidade de vizinhos para calcular a média, o padrão é 5.

•*weights{‘uniform’, ‘distance’}or callable*: O peso de cada vizinho para a média, o padrão é todos com o mesmo peso. 

•*metric{‘nan_euclidean’} or callable*: Métrica das distâncias entre os vizinhos, o padrão é nan_euclidean.

```python
from sklearn.impute import KNNImputer

#knn:
imp = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')

'''Assim, podemos transformar o dataset. Entretanto o KNN é normalmente utilizado em 
dimensão 2, e como uma coluna tem dimensão 1, usamos o .ravel() para o achatamento'''
treino["LotFrontage"] = imp.fit_transform(treino[["LotFrontage"]]).ravel()

print("Temos NaNs na 'LotFrontage'? {}\n".format(treino['LotFrontage'].isnull().any()))

#coluna sem NaNs:
print(treino['LotFrontage'])
```

## **Tipagem dos dados:**

Mesmo com todos os dados dos dois dataframes presentes sendo válidos, as suas colunas ainda estão misturadas em dados numéricos e categóricos; contínuos e discretos; floats, inteiros e strings.

Como todas as colunas dos dois datasets necessitarão ser separadas por tipo, é eficiente juntarmos os dois para manipulação única. 

```python
#Vemos o tamanho de cada df:
print(treino.shape)
print(teste.shape) # o teste só tem 75 colunas porque não tem o preço da casa, que é o que procuramos.
```

#### **Juntando datasets**

É mais lógico juntar os dados para continuar o processamento do que fazer as mesmas coisas duas vezes. 
Para isso criamos um dataframe com todos os dados usando a função .concat():

```python
df_total = pd.concat([treino,teste],ignore_index=True)
print(df_total.shape) #sabendo que: os primeiros 1459 registros vem do dados, e os 1459 últimos vem do teste.
```

## **Colunas numéricas e colunas categóricas:**
Iremos dividir o nosso dataframe total em colunas numéricas e categóricas para trabalhar com cada uma separadamente.

Para isso temos a função [ .select_dtypes](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.select_dtypes.html), que como já diz, seleciona as colunas com o tipo do argumento.

```python
numericos = df_total.select_dtypes(np.number)
col_numer = numericos.columns
print(col_numer)
print("\nSão {} colunas numéricas.".format(len(col_numer)))
```

```python
categoricos = df_total.select_dtypes(exclude= np.number)
col_categ = categoricos.columns
print(col_categ)
print("\nSão {} colunas categóricas.".format(len(col_categ)))
```

## **Processamento dos dados categóricos**

Um dos métodos para vetorizar os dados dessas colunas é o [LabelBinarizer](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelBinarizer.html), mas com apenas uma coluna pelo método one-vs-all. 

Para fazer múltiplas colunas de uma vez podemos utilizar o [.get_dummies](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html) e mudar os parâmetros para colunas, uma vez que o padrão são linhas.

```python
df_preparado = pd.get_dummies(df_total, columns=col_categ)
df_preparado.shape
#são 37 colunas originalmente numéricas, e 39 colunas transformaram-se em 239 colunas geradas pelo .get_dummies.

#Entretetanto,agora temos 2 de cada coluna categórica, a normal e a vetorizada.
# Podemos tirar isso com a drop_duplicates()
print(df_preparado.shape)
df_preparado.drop_duplicates()
df_preparado.head()
```

Agora temos um dataframe único com todas as colunas numéricas. Exceto, é claro, a coluna 'SalePrice'. 

Os primeiros 1459 registros possuem algum valor(vindos do dataset dados), e os outros 1459 possuem NaN(vindos do dataset teste).

```python
#Dados rotulados da 'SalePrice' vindos do df treino
print(df_preparado['SalePrice'].head())
#Dados rotulados da 'SalePrice' vindos do df teste
print(df_preparado['SalePrice'].tail())
```

Como o dataset que vai ser usado para o treinamento é a partição que contém a coluna 'SalePrice', e como nós só juntamos os datasets para questão de uniformidade de OneHot, podemos redividí-los em treino e teste.

```python
#Vamos fazer uma cópia das primeiras 1459 linhas para o df_treino
df_treino=df_preparado.iloc[:1459,:].copy()
#Vamos fazer uma cópia das últimas 1459 linhas para o df_teste
df_teste=df_preparado.iloc[1459:,:].copy()

#Como foi criado uma coluna 'SalePrice' no dataframe de teste quando juntamos os datasets, vamos retirá-la:
df_teste.drop(['SalePrice'],axis=1,inplace=True)
```

## **Treinamento do modelo**

```python
#vamos criar os dois parâmetros que a função fit recebe:
xtreino = df_treino.drop(['SalePrice'],axis=1)
# O segundo parâmetro que indica os rótulos:
ytreino = df_treino['SalePrice']
```

Separa os dados de treino e teste.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_treino, df_teste, test_size = 0.25, random_state=42)
```

Implementa o modelo Multi-Layer Perceptron (MLP), treinando com o algoritmo Backpropagation.

```python
from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor(random_state=1, max_iter=1000)
modelo_treinado = mlp.fit(xtreino, ytreino)

print(modelo_treinado)

expected_y  = y_test
predicted_y = modelo_treinado.predict(xtreino)
```

## **Avaliação do modelo**

Para avaliar a quantidade de erro do nosso modelo podemos utilizar o RMSE (root mean squared error) no dados de treino. Uma vez que não podemos calcular a acurácia nos dados de teste porque não temos os rótulos.

```python
from sklearn.metrics import mean_squared_error
p = modelo_treinado.predict(xtreino)

print("Quantidade de erros: {}".format(np.sqrt(mean_squared_error(ytreino, p))))
```

Com isso, podemos voltar no nosso regressor e ajustar alguns de seus parâmetros manualmente, de forma a diminuir esse erro.

```python
##criamos nosso regressor:
regressor2= MLPRegressor(
                           hidden_layer_sizes=(20,20,20), #três camadas ocultas são modeladas, com 20 neurônios em cada camada
                           activation="relu", #função de ativação para a camada oculta
                           random_state=1, 
                           max_iter=5000).fit(xtreino, ytreino
                            )#aplicamos o regressor no modelo.

modelo_treinado2 = regressor2

p = modelo_treinado2.predict(xtreino)

print("Quantidade de erros: {}".format(np.sqrt(mean_squared_error(ytreino, p))))
```

```python
#Vamos comparar as predições:
predicao1 = modelo_treinado.predict(df_teste)
print("Predição1:",predicao1)
predicao2 = modelo_treinado2.predict(df_teste)
print("Predição2:",predicao2)
```
